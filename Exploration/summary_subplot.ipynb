# 1:Figures和Axes的关系
使用轴创建图形的最简单方法是使用plt.subplots(). 然后我们可以使用 Axes.plot在轴上绘制一些数据。

正如上图所示，一个axes中可以包含整幅图的标题(Title)、坐标轴名称(X or Y axis label)，刻度(major tick)，

刻度值(tick label)、图例(legend)、网格(Grid)、画图时所选用的标记(Markers),可以为圆，三角形，正方形，十字等等…

因此，选取matplotlib画图时，首先创建一个画布Figure,在该画布中可以声明一个或多个Axes, 每一个Axes都是一个独立的一个对象：

```python
import matplotlib.pyplot as plt
import numpy as np
fig,ax = plt.subplots()# creat a figure named fig, an axes named ax # 创建画布，一个AXES种应当只有一个figure.
# I plan to draw 3 lines y1, y2 and y3 in a figure and make 
# some modifications and decorations on the Axes
x = np.linspace(-2*np.pi,2*np.pi,100)
y1 = x**2
y2 = 20*np.sin(x)
y3 = 20*np.cos(3*x)
```

```python
# Fig.1:plot by matlab formats
plt.plot(x,y1)
plt.show()
```

```python
# Fig.2: 1 Figure, 1 Axes
fig1, ax1 = plt.subplots() # Create a figure contained an axes
ax1.plot(x,y1)
plt.show()
```

```python
#Fig.3: 1 Figure, 2 axes
fig2, (axs1,axs2) = plt.subplots(2,1) # Create a figure contained 2 axes #一个画布上分配了2张图 axs1 和 axs2，在画图的时候这两个互不影响
#图的排序是2行一列
axs1.plot(x,y2)
axs2.plot(x,y3)
plt.show()
```

```python
#Fig.3: 1 Figure, 2 axes
fig2, (axs1,axs2) = plt.subplots(1,2) # Create a figure contained 2 axes #一个画布上有2张图，图的排序是1行2列
axs1.plot(x,y2)
axs2.plot(x,y3)
plt.show()
```

# 我们可以将plt的画图方式归结为两类：
1. 面向对象：直接创建画布和坐标轴，并调用其中的方法(object-oriented style)

2. 面向plt：依赖matplotlib.Pyplot直接画图并且改变相应的特征值。

# 1. 面向对象 OO style

```python

fig, ax = plt.subplots(figsize=(13,6)) # Create a figure and an axes.
ax.plot(x, y1, label='Quadratic') # Plot some data on the axes.
ax.plot(x, y2,label='20Sin(x)') # Plot more data on the axes...
ax.plot(x, y3, label='20Cos(3x)') # ... and some more.

ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(13,6)) # Create a figure and an axes.
ax.plot(x, y1, label='Quadratic') # Plot some data on the axes.
ax.plot(x, y2,label='20Sin(x)') # Plot more data on the axes...
ax.plot(x, y3, label='20Cos(3x)') # ... and some more.

# 黄色填充满足条件的区域
ax.fill_between(x, y1, y2, where=(y2 > y1), facecolor='yellow', alpha=0.5) # set the transparency of the area
xmin, xmax, ymin, ymax = ax.axis()
ax.hlines(0,xmin, xmax,ls='-.') # draw a horizontal line
 
 # 使用annotate指令添加箭头添加标记
ax.annotate('y2>y1', xy=(1.2, 10), xytext=(2.2, 25),
arrowprops=dict(facecolor='black', shrink=0.05),fontsize=16)
 

ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(13,6)) # Create a figure and an axes.
ax.plot(x, y1, label='Quadratic') # Plot some data on the axes.
ax.plot(x, y2,label='20Sin(x)') # Plot more data on the axes...
ax.plot(x, y3, label='20Cos(3x)') # ... and some more.
ax.set_axis_on() #设置坐标轴是否显示，off即在画布中不会出现坐标轴及其刻度，只有一幅图像。
ax.set_facecolor('beige') # 设置axes的颜色，可以理解为整个画图的背景颜色。

#设置坐标轴的范围
#ax.set_xlim([xmin,xmax])
#ax.set_ylim([ymin,ymax])
# 也可以直接设置整个axis的范围ax.axis([xmin, xmax, ymin, ymax])

# 设置坐标轴的名称及整幅图的标题
ax.set_xlabel("X-label",fontsize=16,fontstyle='oblique') 
ax.set_ylabel("y-label",fontsize=16,fontstyle='oblique') 
#ax.set_title("Compare")
ax.legend(loc="upper right",shadow="True",fontsize=12,facecolor="lightskyblue")
#ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(13,6)) # Create a figure and an axes.
ax.plot(x, y1, label='Quadratic') # Plot some data on the axes.
ax.plot(x, y2,label='20Sin(x)') # Plot more data on the axes...
ax.plot(x, y3, label='20Cos(3x)') # ... and some more.
ax.set_axis_on() #设置坐标轴是否显示，off即在画布中不会出现坐标轴及其刻度，只有一幅图像。
ax.set_facecolor('beige') # 设置axes的颜色，可以理解为整个画图的背景颜色。

#设置坐标轴的范围
#ax.set_xlim([xmin,xmax])
#ax.set_ylim([ymin,ymax])
# 也可以直接设置整个axis的范围ax.axis([xmin, xmax, ymin, ymax])

# 设置坐标轴的名称及整幅图的标题
ax.set_xlabel("X-label",fontsize=16,fontstyle='oblique') 
ax.set_ylabel("y-label",fontsize=16,fontstyle='oblique') 
#ax.set_title("Compare")
ax.legend(loc="upper right",shadow="True",fontsize=12,facecolor="lightskyblue")
#ax.legend()

 # 使用浅红色显示出y2<y1的3个区域，并且利用text在各区域标注序号
ax.fill_between(x, y1, y2, where=(y2 < y1), facecolor='peachpuff', alpha=0.5)
ax.text(-5.8,20,str(1),fontsize=16,c='darkorchid');ax.text(-2,-8,str(2),fontsize=20,c='darkorchid')
ax.text(5,10,str(3),fontsize=20,c='darkorchid')

plt.show()
```

# 2. 面向plt：依赖matplotlib.Pyplot直接画图并且改变相应的特征值。

```python
# pyplot style
plt.figure(figsize=(13,6))
plt.plot(x, y1, label='Quadratic') # Plot some data on the (implicit) axes.
plt.plot(x, y2, label='20Sin(x)') 
plt.plot(x, y3, label='20Cos(3x)')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot--Pyplot style")
plt.legend()
plt.show()
```
